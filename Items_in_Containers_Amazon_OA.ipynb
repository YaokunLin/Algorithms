```python
input='*|***|**|*****|'
startIndices=[1, 2, 7,7]
endIndices=[6, 15, 14,15]

def solution(input,startIndices,endIndices):
  compList=[]
  startIdx = 0
  endIdx = input.find("|")
  for idx in range(len(input)):
    if input[idx]=="|":
      startIdx,endIdx=endIdx,idx
      if startIdx!= endIdx: compList.append((startIdx,endIdx))

  #return compList
  #compList=becomes [(1,5),(5,8),(8,14)]
  #Time: O(num of elements)
  #Space: O(num of "|")
 

  returnList =[]
  for i in range(len(startIndices)):
    startWindow= startIndices[i]-1 #0 based
    endwindow= endIndices[i]-1 #0 based
    temp=0

    for startComp, endComp in compList:
      if startComp>=startWindow and endComp<=endwindow:
        temp=temp+(endComp-startComp-1)
      if startComp>endwindow: break

    returnList.append(temp)

  return returnList
```

```python
solution(input,startIndices,endIndices)
```
