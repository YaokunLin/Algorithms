```python
def isHoppable( Towers):
  location =[x for x in range(len(Towers))]  #[0,1,2,3,4,5]
  stepsNeededEscape = [ len(Towers)-x for x in location] #[6,5,4,3,2,1]

  def hop(jumpOffPoint):
    if jumpOffPoint <= len(Towers)-1:
      reachDistance=Towers[jumpOffPoint]
      if  reachDistance >= stepsNeededEscape[jumpOffPoint]: return True
      if  reachDistance == 0: return False

    Escaped = False
    jumpDistance = 1
    while jumpDistance <= reachDistance and not Escaped: 
      Escaped = hop(jumpOffPoint+jumpDistance)
      jumpDistance +=1
    return Escaped

  return hop(0)
```

```python
Towers=[4,2,0,0,2,0]
isHoppable(Towers)
```

```python
Towers=[1,0]
isHoppable(Towers)
```

```python
Towers=[3,2,0,0,2,0]
isHoppable(Towers)
```

```python
Towers=[3,2,0,1,2,0]
isHoppable(Towers)
```

```python
def isHoppableDP( Towers):
  location =[x for x in range(len(Towers))]  #[0,1,2,3,4,5]
  stepsNeededEscape = [ len(Towers)-x for x in location] #[6,5,4,3,2,1]
  EscapedMemo=[None] * len(Towers)

  def hop(jumpOffPoint):
    if jumpOffPoint <= len(Towers)-1:
      reachDistance=Towers[jumpOffPoint]
      if  reachDistance >= stepsNeededEscape[jumpOffPoint]: return True
      if  reachDistance == 0: return False

    Escaped = False
    jumpDistance = 1
    while jumpDistance <= reachDistance and not Escaped: 
      Escaped = EscapedMemo[jumpOffPoint+jumpDistance]
      jumpDistance +=1
    return Escaped

  jumppinPoint = len(Towers)-1
  while jumppinPoint >=0:
    EscapedMemo[jumppinPoint ] = hop(jumppinPoint)
    jumppinPoint-=1

  #EscapedMemo =[True,False,False,False,True,False]
  return EscapedMemo[0]
```

```python
Towers=[4,2,0,0,2,0]
isHoppableDP(Towers)
```

```python
Towers=[1,0]
isHoppableDP(Towers)
```

```python
Towers=[3,2,0,0,2,0]
isHoppableDP(Towers)
```

```python
Towers=[3,2,0,1,2,0]
isHoppableDP(Towers)
```
