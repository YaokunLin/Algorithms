```python
def subsets( nums):
  returnList=[]
  fullList=nums[:]
  if fullList==[]: return []
  returnList.append([])

  def removeOne(givenList): #O(2^n)
    if givenList != [] and givenList not in returnList:
      returnList.append(givenList[:])
      for idx in range(len(givenList)):
        temp=givenList[:]
        temp.pop(idx)
        removeOne(temp)
 
  removeOne(fullList)
  return returnList
```

```python
nums=[1,2,3]
subsets(nums)
```
