```python
def solve(M):
  nRow=len(M)
  nCol=len(M[0])
  moves=[(-1,0),(+1,0),(0,-1),(0,+1)]
  q=[]
  q.append((0,0)) # starting point
  seen=set()
  seen.add((0,0))
  minStep=0

  while q:
    OldR,OldC=q.pop()
    minStep+=1
    for rM,cM in moves:
      newR,newC = OldR+rM,OldC+cM
      if 0<=newR<=nRow-1 and 0<=newC<=nCol-1 and (newR,newC) not in seen: 
        if M[newR][newC]==1:
          seen.add((newR,newC))
          q.append((newR,newC))
        if M[newR][newC]==9:
          return newR+newC      
  return -1
```
