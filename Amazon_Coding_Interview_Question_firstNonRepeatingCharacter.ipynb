```python
def firstNonRepeatingCharacter(givenStr):
  seenChar=[]
  charCount=[]

  for char in givenStr:
    if char not in seenChar:
      seenChar.append(char)
      charCount.append(1)
    else:
      charCount[seenChar.index(char)]+=1

  if min(charCount)==1: return seenChar[charCount.index(1)]
  return "_"
```

```python
firstNonRepeatingCharacter("aaabcccdeef")
```

```python
firstNonRepeatingCharacter("abcbad")
```

```python
firstNonRepeatingCharacter("abcabcabc")
```

```python
def firstNonRepeatingCharacter2(givenStr):
  countDict={}

  for char in givenStr:
    if char not in countDict:
      countDict[char]=1
    else:
      countDict[char]+=1

  for char in givenStr:
    if countDict[char] == 1: return char

  return "-_"
```

```python
firstNonRepeatingCharacter2("aaabcccdeef")
```

```python
firstNonRepeatingCharacter2("abcbad")
```

```python
firstNonRepeatingCharacter2("abcabcabc")
```
