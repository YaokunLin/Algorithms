## **Write a function with linear run time O(n)**

```python
def longestConsecutiveChar(givenStr):
  length=1
  maxLen=length
  longestConsecutiveChar=''

  for idx,char in enumerate(givenStr): #O(n)
    if  idx+1 <= (len(givenStr)-1) and (givenStr[idx+1] == char):
      length=length+1
      maxLen=max(maxLen,length)
    else:
      length=1
    if maxLen==length: longestConsecutiveChar=char

  return {longestConsecutiveChar:maxLen}
```
