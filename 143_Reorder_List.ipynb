```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
'''
1234==12+34 fast/slow pointers
34-->43 reverse linked list
12,43 merge

12345=123+45
45-->54
123,54
15243

'''

class Solution:
    def reorderList(self, head: Optional[ListNode]) -> None:
        if head.next==None: return head
        
        dummyHead=ListNode(val=-1,next=head)
        fast,slow=ListNode(val=-1,next=head),ListNode(val=-1,next=head)
        
        #fast/slow pointers
        while fast.next and fast.next.next:
          slow=slow.next
          fast=fast.next.next
        
        secondHead=slow.next
        slow.next=None #break it into 2 lists
        #fast:1,3
        #slow:1,2
        #slow.next=3=secondHead
        
        #fast:1,3,5
        #slow:1,2,3
        #slow.next=4
        
        #reverse linked list
        cur=secondHead
        nxt=cur.next
        while nxt:
          nxtnxt=nxt.next
          nxt.next=cur
          cur=nxt
          nxt=nxtnxt
        secondHead.next=None #it becomes a tail
        #3->4->None
        #None<-3<-4
        #cur=3,nxt=4,nxtnxt=None
        #4.next=3
        #cur=4,nxt=None
        
        #3->4->5->None
        #None<-3<-4<-5
        #cur=3,nxt=4,nxtnxt=5
        #4.next=3
        #cur=4
        #nxt=5
        #nxtnxt=None
        #5.next=4
        #cur=5
        #nxt=None
        
        #merge
        #1->2->None,4->3->None merge
        #123,54 merge
        cur1=dummyHead.next
        cur2=cur
        
        while cur1.next:
          cur1Nxt,cur2Nxt=cur1.next,cur2.next
          cur1.next=cur2
          if cur2.next: cur2.next=cur1Nxt
          cur1=cur1Nxt
          cur2=cur2Nxt
        
        if cur2:cur1.next=cur2
        #cur1=1,cur2=4
        #cur1Nxt,cur2Nxt=2,3
        #1.next=4,4.next=2
        #cur1=2,cur2=3
        
        #cur1=1, cur2=5
        #cur1Nxt,cur2Nxt=2,4
        #1.next=5,5.next=2
        #cur1=2, cur2=4
        #cur1Nxt=3,cur2Nxt=None
        #3.next=4
        #cur1=3, cur2=None
        
        return dummyHead.next
        
          
        
        
          
          
        

        
```
